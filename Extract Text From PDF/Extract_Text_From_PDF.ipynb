```python
#import lib
from os import read
import PyPDF2
#open the pdf
pdf = open('/content/intro to ML.pdf', 'rb')

#read pdf
reader = PyPDF2.PdfFileReader(pdf)
#get number of page
totalpages = reader.numPages

#create list for save text
list_text = []

for number in range(0 , totalpages):
  #get details of page
  page = reader.getPage(number)
  #extract text
  text = page.extractText()
  list_text.append(text)

#print 
for text in list_text:
  print(text)
  print("-------------------------------")



```
